2. Pobreza
El siguiente segmento de código realiza un análisis de componentes principales (PCA) para los indicadores de pobreza en los municipios de Colombia. Los indicadores a usar en el análisis son analfabetismo, inasistencia escolar, alta dependencia económica y fuente de agua mejorada. Las cifras se obtuvieron del siguiente URL: https://github.com/finiterank/mapa_pobreza_colombia/blob/master/data/pobreza.csv

```python
#Inicio leyendo los datos con la libreria cvs. Luego paso toda la información a un array llamado data.
%pylab inline
import csv
from __future__ import division
archivo= open('pobreza.csv', 'rb')
data_csv = csv.reader(archivo)
data=[]
prome=[]
count =-1
cova=[]
variables = array(['analfabetismo','inasistencia escolar', 'alta dependencia económica', 'fuente de agua mejorada'])
for row in data_csv:
    count+=1
    line=[]
    if (count!=0): 
        line.append(row[11])
        line.append(row[13])
        line.append(row[21])
        line.append(row[29])
        if all([(a!='NA' and a!='#DIV/0!') for a in line]):#Corrigo la presencia de posibles inconsistencias.
            for i in range(len(line)):
                line[i]=float(line[i])
            data.append(line)
   
data=array(data)
data_csv =close()

print ''
print 'Finalmente, los datos sobre los que voy a realizar el análisis se encuentran en data.'
print 'La información esta en un array de (1113 X 44), con 4 datos de 1113 municipios.'
print shape(data)

```

```python
#Encuentro para cada columna en data (para cada variable) el valor medio.
for i in range(shape(data)[1]):
    prome.append(mean(data[:,i]))
#Calculo la matriz de desviaciones de la media restando a cada elemento en data la media de su columna
for row in data:
    for i in range(len(row)):
        row[i]=row[i]-prome[i]
#Encuentro la matriz covarianza usando la definición de covarianza.    
count=-1
for k in range(len(prome)**2):
    u=k%len(prome)
    a=0
    if(u==0):
        cova.append([])
        count+=1
    for i in range(len(data[:,0])):
        a+=(data[i,(count)])*(data[i,u])    
    cova[count].append(a/(len(data[:,0])-1))

cova=array(cova)

print ''
print 'Una gráfica del log10 de la matriz covarianza muestra que es simetrica.'
imshow(log10(cova))
```

```python
#Calculo los valores y vectores propios de la matriz covarianza.
eigenvalues, eigenvectors = linalg.eig(cova)

#Realizo la proyeccion de los datos sobre las nuevas bases (eigenvectors)
row_vectors=eigenvectors.T
result=dot(row_vectors,(data.T))

print ''
print 'El resultado de este proceso es un array (4 X 1113).'
print 'Este array muestra los coeficientes de ajuste necesarios para recrear los datos iniciales a partir de una '
print 'combinación lineal de la base.'
print shape(result)
```

```python
#La siguiente gráfica muestra los coeficientes de ajuste para cada vector, para cada grupo de datos (municipio).
for i in range(shape(result)[1]):
    plot (result[:,i],'ko')
xlabel('Vector base')
ylabel('Coeficiente de ajuste')
print ''
print 'Esta gráfica nos muestra que los coeficiente de ajuste para el primer vector varia mas que para los demas.'
print 'Sin embargo, no existe una gran diferencia entre los coeficientes de los primeros 3 vectores.'
```
