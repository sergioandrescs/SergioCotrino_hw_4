2. Pobreza
El siguiente segmento de código realiza un análisis de componentes principales (PCA) para los indicadores de pobreza en los municipios de Colombia. Los indicadores a usar en el análisis son analfabetismo, inasistencia escolar, alta dependencia económica y sin fuente de agua mejorada. Las cifras se obtuvieron del siguiente URL: https://github.com/finiterank/mapa_pobreza_colombia/blob/master/data/pobreza.csv

Para realizar el analisis de datos y el PCA se tomo información de las siguientes fuentes:

JOLLIFFE I.T. Principal Component Analysis.Second Edition. Springer.New York, 2002

```python
#Inicio leyendo los datos con la libreria cvs. Luego paso toda la información a un array llamado data.
%pylab inline
import csv
from __future__ import division
archivo= open('pobreza.csv', 'rb')
data_csv = csv.reader(archivo)
data=[]
prome=[]
count =-1
cova=[]
variables = array(['analfabetismo','inasistencia escolar', 'alta dependencia económica', 'fuente de agua mejorada'])
for row in data_csv:
    count+=1
    line=[]
    if (count!=0): 
        line.append(row[11])
        line.append(row[13])
        line.append(row[21])
        line.append(row[29])
        if all([(a!='NA' and a!='#DIV/0!') for a in line]):#Corrigo la presencia de posibles inconsistencias.
            for i in range(len(line)):
                line[i]=float(line[i])
            data.append(line)
   
data=array(data)
data_csv =close()

print ''
print 'Finalmente, los datos sobre los que voy a realizar el análisis se encuentran en data.'
print 'La información esta en un array de (1113 X 44), con 4 datos de 1113 municipios.'
print shape(data)

```

```python
#Encuentro para cada columna en data (para cada variable) el valor medio.
for i in range(shape(data)[1]):
    prome.append(mean(data[:,i]))
#Calculo la matriz de desviaciones de la media restando a cada elemento en data la media de su columna
for row in data:
    for i in range(len(row)):
        row[i]=row[i]-prome[i]
#Encuentro la matriz covarianza usando la definición de covarianza.    
count=-1
for k in range(len(prome)**2):
    u=k%len(prome)
    a=0
    if(u==0):
        cova.append([])
        count+=1
    for i in range(len(data[:,0])):
        a+=(data[i,(count)])*(data[i,u])    
    cova[count].append(a/(len(data[:,0])-1))

cova=array(cova)

print ''
print 'Una gráfica del log10 de la matriz covarianza muestra que es simetrica.'
imshow(log10(cova))
```

```python
#Calculo los valores y vectores propios de la matriz covarianza.
eigenvalues, eigenvectors = linalg.eig(cova)

#Realizo la proyeccion de los datos sobre las nuevas bases (eigenvectors)
row_vectors=eigenvectors.T
result=dot(row_vectors,(data.T))

print ''
print 'El resultado de este proceso es un array (4 X 1113).'
print 'Este array muestra los coeficientes de ajuste necesarios para recrear los datos iniciales a partir de una '
print 'combinación lineal de la base.'
print shape(result)
```

```python
#La siguiente gráfica muestra los coeficientes de ajuste para cada vector, para cada grupo de datos (municipio).
for i in range(shape(result)[1]):
    plot (result[:,i],'ko')
xlabel('Vector base')
ylabel('Coeficiente de ajuste')
print ''
print 'Esta gráfica nos muestra que los coeficiente de ajuste para el primer vector varia mas que para los demas.'
print 'Sin embargo, no existe una gran diferencia entre los coeficientes de los primeros 3 vectores.'
print 'Entonces los dos primeros componentes principales se muestran como buenas candidatos para'
print 'reducir la dimensionalidad.'
```

```python
#Si se desean formar dos clusters
clust=2
cl=[]
#Linea para asignar los colores del cluster
color ='rgbymkc'
count=0
#Crea una nueva lista con lista de 3 elementos: coordenada x, coordenada y y color
for i in range(len(result[0,:])):
    cl.append([])
    cl[count].append(result[0,i])
    cl[count].append(result[1,i])
    cl[count].append(color[i%clust])
    count+=1   
dif = -1
counter = 0
while  dif !=0 and counter <10: 
    #Calcula el centro de las distribuciones
    counter +=1
    center=[]
    dif = 0
    for k in range(clust):
        c_x=0
        c_y=0
        count=0
        for i in range(len(result[0,:])):     
            if(cl[i][2]==color[k]):
                c_x+=cl[i][0]
                c_y+=cl[i][1]
                count +=1
        if(count>0):
            center.append([c_x/count,c_y/count]) 

    #realiza el analisis de distancia a cada centroide creado un lista de distancias y
    #encontrando el valor mas pequeño.
    for i in range(len(result[0,:])):
        dst=[]
        for k in range(clust):
            s=sqrt(((cl[i][0]-center[k][0])**2)+(cl[i][1]-center[k][1])**2)
            dst.append(s)
            
        if (cl[i][2]!= color[where(dst==min(dst))[0][0]]):
            dif+=1
        cl[i][2]= color[where(dst==min(dst))[0][0]]  
    
for i in range(len(result[0,:])):
    scatter(cl[i][0],cl[i][1], c=cl[i][2]) 
for k in range(clust): 
    scatter(center[k][0],center[k][1], c=color[k], s=120)
xlabel('PC1')
ylabel('PC2')
grid()
print 'Una separacion en dos clusters puede mostrar las ciudades con y sin acceso a fuentes de agua mejorada.'
```
